#### 2.	Explain the various techniques to store a network in mathematical representation (200-250 words)
>Visual representation of a graph is an easy task, however storing it in an optimal way is a challenge. Various techniques exist to store a network in mathematical representation and each has their pros and cons. Mainly three types of techniques are quite prominent for this: Adjacency Matrix, Adjacency List and Edge List.

><u>Adjacency Matrix</u> – As the name suggests, the storage is done in the form of a matrix, where the values of the matrix represent whether two nodes are adjacent to each other or not(share the edges or have any ties).

>In case of a social graph half of the matrix is filled with zeros: which represents no link or mostly sparse, since in real world problems everyone is not connected to every other leading to wastage of space. If the graph is undirected then one half the matrix is mirroring the second half, which again is wastage of space. However the good part of this representation is the fast access since array access takes constant time.

><u>Adjacency List</u> – Here instead of having the matrix to represent the links we represent the links in the form of a list. Each node has its own adjacency list of the nodes connected to it. It utilizes less space as compared to the adjacency matrix, however it is not easy to cross check if a node is adjacent to other, as we might have to iterate over the whole adjacency list.

>In this case also if the graph is undirected, then both the nodes will have entry of each other in their adjacency list.

><u>Edge List</u> -  This is in the list of the edges of a graph. It is optimal in terms of space as compared to the other two methods. It is a list of tuples, which is represented as (u, v) denoting that the node u is connected to node v via that edge.

#### a.	Loads the necessary libraries to run (e.g., networkx, community, infomap, etc.)

```python
import networkx as nx
#import infomap as im
from community import community_louvain
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
import gzip
import pprint
import json
from collections import Counter
```

#### b.	Loads the “GamersGate” twitter dataset

```python
users = set()
p_edge_list = []

# Iterate over the file
with gzip.open('gamergate.json.gz','rb') as f:
    for line in f:
        # Conver to the JSON format
        status = json.loads(line)
        # Add the user entry
        users.add(status['user']['id_str'])
        if not status['text'].startswith("RT @"):
            # Iterate over all the user mentions
            for user_mentions in status['entities']['user_mentions']:
                p_edge_list.append((status['user']['id_str'], user_mentions['id_str']))
        
edge_list = []
for edge_pair in p_edge_list:
    # First user will always be in there since he shall be the person who tweets
    if edge_pair[1] in users:
        edge_list.append(edge_pair)
```

```python
print(len(users))
print(len(edge_list))
```

#### c.	Builds a mention graph from users mentioning (@) other user names in their tweets

```python
G = nx.Graph()
G.add_edges_from(edge_list)
```

#### e.	Calculates the following network statistics:  node degree distribution, number of connected components, network density, number of nodes and edges.

```python
print(nx.info(G))
```

```python
# Diameter is infinity since too many disconnected components
# print('Diameter', nx.diameter(G))
print('Components', nx.number_connected_components(G))
print('Density',nx.density(G))
print('Degree', G.degree())
```

```python
# Cross check min value should be >= 1
print(min(dict(G.degree).values()))
```

#### Node degree distribution

```python
dc = Counter(dict(G.degree).values())
x = []
y = []
for degree, count in sorted(dc.items(), key=lambda x:x[0]):
    x.append(degree)
    y.append(count)

plt.plot(x, y,'b+-')
plt.xlabel('Degree')
plt.ylabel('Nodes')
plt.title('Degree distribution of network')
plt.show()
```

```python
# Maximum degree possible
print(max(G.degree(),key = lambda x:x[1]))
```

```python
plt.loglog(x, y,'go-')
plt.xlabel('log(Degrees)')
plt.ylabel('log(Nodes)')
plt.title('loglog Plot of Degree Distribution')
plt.show()
```

#### d.	Calculates the following network centrality measures: degree centrality, node clustering coefficient, eigenvector centrality, PageRank.

```python
deg_cen = nx.degree_centrality(G)
print('Degree centrality:', deg_cen,'\n')
```

#### f.	Calculate the network community structure with an algorithm of your choice, appropriate to the size of the network.
#### g.	Plot the following distributions: node degree distribution, community-size distribution

```python
part = community_louvain.best_partition(G)
size = len(set(part.values()))
print('No. Communities', size)
```

```python
community_nodes = Counter(part.values())
x = []
y = []

for comm, nodes in sorted(community_nodes.items(), key=lambda x:x[0]):
    x.append(comm)
    y.append(nodes)
```

```python
plt.plot(x, y,'bo-')
plt.xlabel('Community ID')
plt.ylabel('Number of Nodes')
plt.title('Node distribution of communities/Community size')
plt.show()
```
