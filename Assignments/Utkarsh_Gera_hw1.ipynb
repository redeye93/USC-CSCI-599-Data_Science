#### 3. Explain, in simple terms, how to build a Decision Tree, including all the steps of the algorithm (200-250 words).

>Decision trees is one of the supervised learning algorithms and as the name suggests the model formed after learning looks like a tree. These are constructed recursively from the training data using the top down approach in which the features are sequentially selected. The decision tree requires discrete attributes but in real life attributes can be continuous and hence they need to converted into discrete ones by creation of intervals where each interval will now be an attribute value. The steps of the algorithm are as follows.

>One of the decision tree algorithm ID3 receives the dataset, attributes that lead to the formation of this node and the attribute set. 
Once the algorithm starts its main job is to find the attribute that best leads to the formation of the branches that have most pure dataset classified to it (as it follows a greedy approach). 
If the algorithm receives a request, it will first create a root node then check if all the examples provided to it are of a particular class then assign that node that particular class or if there is no attribute in attribute set to further create branches then assign the node with the most common value of the target attribute received by it. However if it doesn’t fall in any of the above cases then it tries to find out the best attribute among the attribute set that would create most pure branches meaning subset with same class attribute value. Our **main objective** is to classify the data to particular label, hence getting a pure branch of a label makes sense.

>Once it finds the attribute that leads to the most pure branches it recursively calls itself for each of the subbranches that now have disjoint subsets of the actual data set, that was passed to its parent node, created by the branches. It will keep on the recursive calls till it reaches its base case. The tree structure thus forms will have tree paths that will be mutually exclusive and exhaustive as every instance sill be covered by the tree paths which can also be termed as a single rule. 

>One of the ways of calculating which attribute leads to pure branch is Information gain method which measures the entropy.

>Here we consider two classes: one being + and other being -. The more pure the subset, less will be its entropy. If the subset contains all the instances of the same class it’s entropy shall be zero. We calculate the information one gains if we choose a particular attribute to create a branch for that node which is nothing but the change in the entropy. The more the information gain, the more closer we are in achieving a pure subset.

>**IG = H(T) - H(T|a). where H(T) is the entropy before the split and H(T|a) is the entropy after the split on attribute a.  
H(T) = - sum over all classes(P(x<sub>i</sub>) * log(P(x<sub>i</sub>))) where P(x<sub>i</sub>) is the probability of class i.**

#### 4. Describe the concept of Performance Evaluation in Supervised Learning, including types of errors, performance metrics and methods to evaluate performance (150-200 words). 

>After the classifier is constructed we need a insight on its performance before it can be used on real life applications. In order to evaluate a model we need a testing set that has been labelled, but we only have the training set that has labels. Hence we will use a small portion of our training set for testing and the other part for training also known as **train-test split**. We will hid the labels of the testing set and feed it to our model and then later on compare the results with the ground truths or the masked labels.
<br>
**KFold Cross Validation** - Since our accuracy is greatly impacted by the way data is arranged in our training and testing set, we can get a very high variance estimate. In order to overcome this limitation we can use the KFold Cross Validation method. In this the training set is shuffled and divided into k equal sized sets. The algorithm is run k times with 1 of the k sets as testing set and rest k-1 as training set and each time we use different training set so all the k sets have at least one chance of being used  as a testing set. Then the average of the algorithm over the k rounds is taken as its performance.

##### Quantifying the performance<br>

>**Regression(continuous labels) Error Metrics**<br>
The Labels are not discrete so we directly use statistics of measure the performance. <br>
Mean Absolute Error - 1/N x SUM(all test samples)(|y-y_hypothesis|)  
Mean Squared Error - 1/N x SUM(all test samples)(|y-y_hypothesis|<sup>2</sup>)    
Root Mean Squared Error - SQRT(1/N x SUM(all test samples)(|y-y_hypothesis|<sup>2</sup>)) <br><br>



>**Classification(discrete labels) Error Metrics**<br>
The labels are discreate so we can directly measure the accuracy by dividing the number of correctly predicted labels by the total number of predictions. Some people prefer to use the error rate which is nothing but 1-accuracy. <br>



>However accuracy alone is not a good measure of the performance. In some scenario we are interested in a particular class like detection of spam or not(or we could positive class is spam and negative if not). Here we are mostly interested whether our model can correctly segregated the spam mails from others and hence our focus is mostly on the spam mail class. In case we only had one spam and our classifier classified every mail as good mail then we will get a 99% accuracy in this case. This becomes useless. In such cases we use precision or recall which help us know how precise and complete is our classifier for a class.

>Before we proceed to the precision we need to know certain terms. The first diagram on the left explains that the terms that were actually positive and we also classified positive by model are True Positive, the terms that were actually positive and classified falsely are known as False Negative. The terms that were classified as negative and were actually negative are True Negative and the terms that were negative and were classified as positive are False Positive.
<br>
Precision is the number of correctly classified positive examples divided by the total number of examples that are classified positive. Recall r is the number of correctly classified positive examples divided by the total number of actual positive examples in the test set. However it is hard to compare classifiers based on these two measures since one can be high and other can be low. 


 <br>
>We can also use F measure that uses both the precision and recall to calculate their harmonic mean. In terms of graphs we use the following: <br>

AUC - Area Under the Curve for ROC. The more AUC the more decisive the classifier is.

d.	Implements an engineered feature (or features) of your own design. You are encouraged to get creative in designing this feature.<br>

f.	Splits the dataset in training and testing sets using KFold Cross Validation (use K=10)<br>
g.	Trains and Tests the performance of the classifier in identifying the sentiment of the tweet (using the ‘airline_sentiment’ column of Tweets.csv).<br>
h.	Computes and outputs the accuracy, precision, recall, and F1 score for each of the 10 folds of cross validation.<br>
i.	Computes and outputs the average accuracy, precision, recall, and F1 score across the 10 folds.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.model_selection import KFold
from IPython.display import display 

def classifier(x, y):
    n_folds = 10
    kf = KFold(n_splits=n_folds)
    kfold_gen = kf.split(x, y)
    dt = DecisionTreeClassifier()
    sum_accuracy, sum_f1, sum_recall, sum_precision  = 0, 0, 0, 0
    print('Triggering Decision Tree Classifier')
    counter = 0

    for each_fold in kfold_gen:
        dt.fit(x[each_fold[0]], y[each_fold[0]])
        y_pred = dt.predict(x[each_fold[1]])

        counter += 1

        accuracy = accuracy_score(y[each_fold[1]], y_pred)
        f1 = f1_score(y[each_fold[1]], y_pred, average='macro')
        precision = precision_score(y[each_fold[1]], y_pred, average='macro')
        recall = recall_score(y[each_fold[1]], y_pred, average='macro')

        sum_accuracy += accuracy
        sum_f1 += f1
        sum_recall += recall
        sum_precision += precision

        print("Calculations for the Fold number {}".format(counter))
        print("f1 = ",f1)
        print("accuracy = ",accuracy)
        print("recall = ", recall)
        print("precision = ", precision)

        print("Ending of the fold number= ",counter)
        print("#########################################")

    print("#########################################")
    print("Mean accuracy of Decision Tree = ",sum_accuracy/n_folds)
    print("Mean f1-score of Decision Tree = ",sum_f1/n_folds)
    print("Mean recall of Decision Tree = ",sum_recall/n_folds)
    print("Mean precision of Decision Tree = ",sum_precision/n_folds)
    print("#########################################")
```

> Tweet with no stop words

```python
classifier(X1,y1)
```
